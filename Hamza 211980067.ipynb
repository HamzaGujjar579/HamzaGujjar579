## df.query()
###### This function is used to query the DataFrame based on an expression. An expression can
######  be as simple as a single condition and as complex as a combination of multiple conditions.

```python
df.query("Quantities > 92")
```

## df.loc()
###### This function is a property of DataFrame that returns the group of rows and columns
######  identified by their labels or names.

```python
df.loc[100,"Sales_Manager"]
```

## df.unique()
###### This function returns the list of unique values in a column or series. Instead of applying on
###### complete DataFrame, it works only on the selected single column.

```python
df["Sales_Manager"].unique()
```

## df.nunique()
##### The nunique() method returns the number of unique values for each column. By specifying the column axis ( axis='columns' ), the nunique() method searches column-wise and returns the number of unique values for each row.

```python
df["Quantity"].nunique()
```

## df.isnull()
##### The isnull() method returns a DataFrame object where all the values are replaced with a Boolean value True for NULL values, and otherwise False.

```python
df["Sales_Manager"].isnull()
```

## df.fillna()
##### The fillna() method replaces the NULL values with a specified value. The fillna() method returns a new DataFrame object unless the inplace parameter is set to True , in that case the fillna() method does the replacing in the original DataFrame instead.

```python
df2 = df.fillna("none")
print(df2)
```

## df.sort_values()
##### Pandas sort_values() function sorts a data frame in Ascending or Descending order of passed Column

```python
df.sort_values(by = "Quantity" ,ascending=False , inplace=False)
```

## df.value_counts()
##### When you use value_counts on a dataframe, it will count the number of records for every combination of unique values for every column.

```python
df["Quantity"].value_counts()
```

## df.nlargest()
##### Pandas nlargest() method is used to get n largest values from a data frame or a series.



```python
df.nlargest(10,"Quantity")
```

## df.nsmallest()
##### The nsmallest() function is used to get the first n rows ordered by columns in ascending order. This method is equivalent to df. sort_values(columns, ascending=True). head(n), but more performant.

```python
df.nsmallest(10,"Quantity")
```

## df.copy()
##### The copy() method returns a copy of the DataFrame.

##### By default, the copy is a "deep copy" meaning that any changes made in the original DataFrame will NOT be reflected in the copy.

```python
df2 = df.copy()
df2
```

## df.rename()
##### The rename() method allows you to change the row indexes, and the columns labels.

```python
df.rename({"0" : "10"} ,axis=0)
df
```

## df.where()
##### DataFrame. where() function is similar to if-then/if else that is used to check the one or multiple conditions of an expression in DataFrame and replace with another value when the condition becomes False.

```python
df.where(df["Quantities"] > 92)

df
```

## df.drop()
##### The . drop() method is a built-in function in Pandas that allows you to remove one or more rows or columns from a DataFrame. It returns a new DataFrame with the specified rows or columns removed and does not modify the original DataFrame in place, unless you set the "inplace" parameter to True.

```python
df2 = df.drop(2,axis=0)
df2
```
